```python
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates


df = pd.read_csv("analysis/data/forest_landscape_restoration_country_tab.tsv", sep ='\t')
# print(df.head())
# plt.figure()

fig, ax = plt.subplots(figsize=(20,10))
df.set_index('year_month', inplace=True)

monthly = df.groupby('year_month').size()
# print(np.sum(monthly))
# print(np.mean(monthly))
#group data by product and display sales as line chart
df.groupby('year_month').size().plot(legend=True, ax=ax, label='Forest and landscape restoration', x_compat=True).legend(fontsize=14)
# print(df.shape)
xx = int(df.shape[0]/96)
ax.axhline(y = xx, color = 'b', linestyle = '--', label='Average tweets mention over the 8 years')
labels = ["Forest and landscape restoration", "Average monthly tweet mentions over the 8 years"]
handles, _ = ax.get_legend_handles_labels()
plt.legend(handles = handles[0:], labels = labels, loc='center left', fontsize = 16)

ax.set_xlabel('Year', fontweight ='bold', fontsize = 16)
ax.set_ylabel('Number of tweet mentions', fontweight ='bold', fontsize = 16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.tick_params(axis='both', which='minor', labelsize=12)
# ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.savefig("forest_landscape_restoration_all.jpg", dpi=300)
```

```python
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates


df1 = pd.read_csv("analysis/data/ecological_restoration_country_tab.tsv", sep ='\t')
df1['Date'] = pd.to_datetime(df1['year_month']).dt.strftime('%Y-%m')
df1.set_index('Date', inplace=True)

df2 = pd.read_csv("analysis/data/ecosystem_restoration_country_tab.tsv", sep ='\t')
df2['Date'] = pd.to_datetime(df2['year_month']).dt.strftime('%Y-%m')
df2.set_index('Date', inplace=True)

df3 = pd.read_csv("analysis/data/landscape_restoration_country_tab.tsv", sep ='\t')
df3['Date'] = pd.to_datetime(df3['year_month']).dt.strftime('%Y-%m')
df3.set_index('Date', inplace=True)

df4 = pd.read_csv("analysis/data/forest_landscape_restoration_country_tab.tsv", sep ='\t')
df4['Date'] = pd.to_datetime(df4['year_month']).dt.strftime('%Y-%m')
df4.set_index('Date', inplace=True)


fig, ax = plt.subplots(figsize=(20,10))

#group data by product and display sales as line chart
xx1 = df1.groupby('Date').size().to_frame('size')
xx1['scale'] = xx1['size']/xx1['size'].sum()*100
xx1.plot( y = 'scale',legend=True, ax=ax, label='Ecological restoration', x_compat=False).legend(fontsize=14)

xx2 = df2.groupby('Date').size().to_frame('size')
xx2['scale'] = xx2['size']/xx2['size'].sum()*100
xx2.plot( y = 'scale',legend=True, ax=ax, label='Ecosystem restoration', x_compat=False).legend(fontsize=14)

xx3 = df3.groupby('Date').size().to_frame('size')
xx3['scale'] = xx3['size']/xx3['size'].sum()*100
xx3.plot( y = 'scale',legend=True, ax=ax, label='Landscape restoration', x_compat=False).legend(fontsize=14)

xx4 = df4.groupby('Date').size().to_frame('size')
xx4['scale'] = xx4['size']/xx4['size'].sum()*100
xx4.plot(y = 'scale',legend=True, ax=ax, label='Forest and landscape restoration', x_compat=False).legend(fontsize=14)

ax.set_xlabel('Year', fontweight ='bold', fontsize = 16)
ax.set_ylabel('Percentage of tweet mentions', fontweight ='bold', fontsize = 16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.tick_params(axis='both', which='minor', labelsize=12)
# ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.savefig("tweet_mentions_all_percent.jpg")
```
